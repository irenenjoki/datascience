```python
def process_fees_payment(paid_amount):
    semester_fee = 50000

    if paid_amount < 0:
        return "Invalid input. Please enter a positive value for fees paid."
    elif paid_amount == semester_fee:
        return "No balance"
    elif paid_amount < semester_fee:
        return f"The remaining balance is {semester_fee - paid_amount} Ksh."
    else:
        credit_amount = paid_amount - semester_fee
        return f"The amount to be credited for the next semester is {credit_amount} Ksh."

def main():
    try:
        fees_paid = float(input("Enter the fees paid by the student: "))
        result = process_fees_payment(fees_paid)
        print(result)
    except ValueError:
        print("Invalid input. Please enter a valid numeric value.")

if __name__ == "__main__":
    main()



```

```python
import math


def calculate_triangle_area(a, b, c):
    # Calculate semi-perimeter
    s = (a + b + c) / 2

    # Heron's formula for triangle area
    triangle_area = math.sqrt(s * (s - a) * (s - b) * (s - c))

    return triangle_area


def main():
    try:
        # Prompt user for triangle dimensions
        side_a = float(input("Enter the length of side A of the triangle: "))
        side_b = float(input("Enter the length of side B of the triangle: "))
        side_c = float(input("Enter the length of side C of the triangle: "))

        # Check if the input values form a valid triangle
        if side_a <= 0 or side_b <= 0 or side_c <= 0 or side_a + side_b <= side_c or side_a + side_c <= side_b or side_b + side_c <= side_a:
            print("Invalid input. The input values do not form a valid triangle.")
        else:
            # Calculate and display the triangle area
            triangle_area = calculate_triangle_area(side_a, side_b, side_c)
            print(f"The area of the triangle is approximately {triangle_area:.3f} square units.")

    except ValueError:
        print("Invalid input. Please enter valid numeric values.")


if __name__ == "__main__":
    main()
```

```python
def calculate_tiles_required(length, width, tile_area=0.36):
    lab_area = length * width
    tiles_required = lab_area / tile_area
    return tiles_required


def main():
    try:
        length = float(input("Enter the length of the lab in meters: "))
        width = float(input("Enter the width of the lab in meters: "))

        if length <= 0 or width <= 0:
            print("Please enter valid positive values for length and width.")
        else:
            tiles_required = calculate_tiles_required(length, width)
            print(f"The number of tiles required is approximately {tiles_required:.2f}.")

    except ValueError:
        print("Invalid input. Please enter valid numeric values.")


if __name__ == "__main__":
    main()
```

```python
def calculate_net_amount(amount_spent):
    if 10000 <= amount_spent <= 14999:
        discount_rate = 0.10
    elif 15000 <= amount_spent <= 24999:
        discount_rate = 0.15
    elif 25000 <= amount_spent <= 29999:
        discount_rate = 0.20
    elif amount_spent >= 30000:
        discount_rate = 0.25
    else:
        discount_rate = 0.0  # No discount for amounts below 10000

    net_amount = amount_spent - (discount_rate * amount_spent)
    return net_amount


def main():
    try:
        amount_spent = float(input("Enter the amount spent in Kshs: "))

        if amount_spent < 0:
            print("Please enter a valid positive amount.")
        else:
            net_amount = calculate_net_amount(amount_spent)
            print(f"The net amount paid after discount is: {net_amount:.2f} Kshs")

    except ValueError:
        print("Invalid input. Please enter a valid numeric value.")


if __name__ == "__main__":
    main()
```

```python
def calculate_salary(hours_worked, hourly_rate=10, lunch_allowance=200, tax_rate=0.3):
    basic_salary = hours_worked * hourly_rate
    gross_salary = basic_salary + lunch_allowance
    income_tax = gross_salary * tax_rate
    net_pay = gross_salary - income_tax
    return basic_salary, gross_salary, income_tax, net_pay


def convert_currency(amount, currency):
    if currency.upper() == 'KSH':
        exchange_rate = 102
        equivalent_amount = amount / exchange_rate
        return equivalent_amount, 'USD'
    elif currency.upper() == 'USD':
        exchange_rate = 1 / 102
        equivalent_amount = amount * exchange_rate
        return equivalent_amount, 'KSH'
    else:
        return None


def main_salary_program():
    try:
        hours_worked = float(input("Enter the number of hours worked: "))

        if hours_worked < 0:
            print("Please enter a valid positive number of hours.")
        else:
            basic, gross, tax, net = calculate_salary(hours_worked)
            print(f"Basic Salary: {basic:.2f}")
            print(f"Gross Salary: {gross:.2f}")
            print(f"Income Tax: {tax:.2f}")
            print(f"Net Pay: {net:.2f}")

    except ValueError:
        print("Invalid input. Please enter a valid numeric value.")


def main_currency_conversion_program():
    try:
        amount = float(input("Enter the amount: "))
        currency = input("Enter the currency (Ksh or USD): ")

        converted_amount, target_currency = convert_currency(amount, currency)

        if converted_amount is not None:
            print(f"The equivalent amount is {converted_amount:.2f} {target_currency}.")
        else:
            print("Invalid currency. Please enter either 'Ksh' or 'USD'.")

    except ValueError:
        print("Invalid input. Please enter a valid numeric value.")


if __name__ == "__main__":
    print("Choose an option:")
    print("1. Calculate Salary")
    print("2. Currency Conversion")

    choice = input("Enter the option number: ")

    if choice == '1':
        main_salary_program()
    elif choice == '2':
        main_currency_conversion_program()
    else:
        print("Invalid choice. Please enter either '1' or '2'.")
```

```python
def celsius_to_fahrenheit(celsius):
    return (celsius * 9 / 5) + 32


def fahrenheit_to_celsius(fahrenheit):
    return (fahrenheit - 32) * 5 / 9


def main_temperature_conversion():
    try:
        choice = input("Choose conversion type (C to F or F to C): ").upper()

        if choice == 'C':
            celsius = float(input("Enter temperature in Celsius: "))
            fahrenheit = celsius_to_fahrenheit(celsius)
            print(f"{celsius}°C is equal to {fahrenheit:.2f}°F.")
        elif choice == 'F':
            fahrenheit = float(input("Enter temperature in Fahrenheit: "))
            celsius = fahrenheit_to_celsius(fahrenheit)
            print(f"{fahrenheit}°F is equal to {celsius:.2f}°C.")
        else:
            print("Invalid choice. Please enter 'C' or 'F'.")

    except ValueError:
        print("Invalid input. Please enter a valid numeric value.")


if __name__ == "__main__":
    main_temperature_conversion()
```

```python
def calculate_bmi(weight, height):
    bmi = weight / (height ** 2)
    return bmi


def categorize_bmi(bmi):
    if bmi < 20.9:
        return "Underweight"
    elif 21.0 <= bmi <= 35.9:
        return "Normal"
    elif 36.0 <= bmi <= 45.9:
        return "Overweight"
    else:
        return "Obese"


def main():
    try:
        weight = float(input("Enter your weight in kilograms: "))
        height = float(input("Enter your height in meters: "))

        if weight <= 0 or height <= 0:
            print("Please enter valid positive values for weight and height.")
        else:
            bmi = calculate_bmi(weight, height)
            category = categorize_bmi(bmi)
            print(f"Your BMI is {bmi:.2f}, and you are categorized as {category}.")

    except ValueError:
        print("Invalid input. Please enter valid numeric values.")


if __name__ == "__main__":
    main()
```
